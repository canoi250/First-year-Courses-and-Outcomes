```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
Grades = pd.read_excel("STEM Outcomes Grades 2002 - 2018 (anon).xlsx")
Major = pd.read_excel("STEM Outcomes Individual Data 2002 - 2018 (anon).xlsx")
MajorTitles = pd.read_excel("Majors Titles and CIP Codes.xlsx")
```

```python
Grades
```

```python
Major
```

```python
MajorTitles
```

```python
#1) I merged the two data tables on the Student ID key and keep only those records with IDs in both data sets. I
#tried doing it in two different ways and got similar results. 
df_merge=pd.merge(Major, Grades, on='Student ID', how='left')
```

```python
df_merge
```

```python
#2) I picked people who did Politics major and generated a data table for people who do Politics as their first or 
#second or third major and contains all the information on their student ID, courses they take and grades they get. 
df_pol=df_merge[(df_merge["Major 1"]=="POL")|(df_merge["Major 2"]=="POL")|(df_merge["Major 3"]=="POL")]
```

```python
df_pol
```

```python
#We then create a list that contains a list of the courses that they took starting from the one that was the most 
#popular with the data frame classcount. We also will ignore Encounters in the graph since it is suggested in the 
#question. 
top_courses = df_pol["Course Title"].sort_values(ascending=False)
```

```python
classcount=top_courses.value_counts().drop("Encounters")
classcount
```

```python
#We then create a bar plot with the 20 courses taken the most by Politics major using function sort_values(). I set the
#color to be yellow and the edgecolor to be black and alpha to be 0.5. 
courses = classcount.sort_values(ascending=False)[:20].plot(kind="bar", color="yellow",alpha=0.5, edgecolor='black')
courses.set_xlabel("Courses")
courses.set_ylabel("Frequency")
courses.set_title("Most Preferred Courses for Politics Majors ")
courses
```

```python
#3)We pick General Chemistry out of the classes shown in the plot excluding Encounters, which is the course all 
#first-year students have to take. 
```

```python
#We create a data table that had all the grades people who have Politics either as their first or second or third major
#get in General Chemistry using the loc() function. 
merge2 = (df_merge['Major 1'] == 'POL') | (df_merge['Major 2'] == 'POL') | (df_merge['Major 3'] == 'POL')
df_merge1 = df_merge.loc[(df_merge['Course Title'] == 'General Chemistry') & merge2, ['Grade', 'Course Title', 'Major 1', 'Major 2', 'Major 3']]
df_merge1
```

```python
#We then create a histogram with the grade distribution of Politics majors taking General Chemistry using the plot()
# function. I set the color to be pink and the edgecolor to be black and alpha to be 0.5 and set Density to be true to
#create normal distribution. 
df_merge1['Grade'].plot(kind='hist', alpha=0.5, color='pink', edgecolor='black', density=True)
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Grade Distribution of General Chemistry Grades for Politics Majors')
```

```python
df_merge1['Grade'].describe()
```

```python
#We create a data table that had all the grades people who do not have Politics neither as their first or second or 
#third major get in General Chemistry using the loc() function. 
merge2 = (df_merge['Major 1'] != 'POL') | (df_merge['Major 2'] != 'POL') | (df_merge['Major 3'] != 'POL')
df_merge3 = df_merge.loc[(df_merge['Course Title'] == 'General Chemistry') & merge2, ['Grade', 'Course Title', 'Major 1', 'Major 2', 'Major 3']]
df_merge3
```

```python
#We then create a histogram with the grade distribution of non-Politics majors taking General Chemistry using the 
#plot() function. I set the color to be orange and the edgecolor to be black and alpha to be 0.5 and set Density to 
#be true to create normal distribution. 
df_merge3['Grade'].plot(kind='hist', alpha=0.5, color='orange', edgecolor='black', density=True)
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.title('Grade Distribution of General Chemistry Grades for non-Politics Majors')
```

```python
df_merge3['Grade'].describe()
```

```python
#The histograms in question 2 and 3 about Politics Majors and non-Politics majors taking General Chemistry show some 
#differences: for example, the graph about Politics Majors has higher standard deviation as more letter grades are 
#represented. The mean grade seems to be quite similar and there is similar frequency of A scores. There is more variation
#for Politics majors as the grades seem to vary more as for non-Politics majors they don't. 
```
